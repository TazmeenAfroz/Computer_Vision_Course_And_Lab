# 🧪 Three Training Variants
1) **Scratch CNN** (train everything)
2) **ResNet50 Feature Extraction** (freeze base)
3) **ResNet50 Fine-tune Last Block** (unfreeze conv5_x only)

Each variant reuses the same folder dataset (`dataset/`).

````python
# ============================================
# HELPERS: DATA PIPELINES + MODELS
# ============================================

data_dir = 'data/images'  # CHANGE as needed
IMG_SIZE = (224, 224)
BATCH_SIZE = 2
SEED = 42

# ---------- DATA LOADERS ----------

def make_raw_datasets():
    train_raw = tf.keras.utils.image_dataset_from_directory(
        data_dir,
        validation_split=0.25,
        subset='training',
        seed=SEED,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode='int'
    )
    val_raw = tf.keras.utils.image_dataset_from_directory(
        data_dir,
        validation_split=0.25,
        subset='validation',
        seed=SEED,
        image_size=IMG_SIZE,
        batch_size=BATCH_SIZE,
        label_mode='int'
    )
    return train_raw, val_raw

# Normalizers
rescale = Rescaling(1./255)
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess


def make_scratch_ds():
    train_raw, val_raw = make_raw_datasets()
    class_names = train_raw.class_names
    train = train_raw.map(lambda x, y: (rescale(x), y))
    val   = val_raw.map(lambda x, y: (rescale(x), y))
    return train, val, class_names


def make_resnet_ds():
    train_raw, val_raw = make_raw_datasets()
    class_names = train_raw.class_names
    train = train_raw.map(lambda x, y: (resnet_preprocess(x), y))
    val   = val_raw.map(lambda x, y: (resnet_preprocess(x), y))
    return train, val, class_names

# ---------- MODELS ----------

def build_scratch_cnn(num_classes):
    return Sequential([
        Conv2D(32, (3,3), activation='relu', padding='same', input_shape=(*IMG_SIZE, 3)),
        MaxPooling2D((2,2)),
        Conv2D(64, (3,3), activation='relu', padding='same'),
        MaxPooling2D((2,2)),
        Conv2D(128, (3,3), activation='relu', padding='same'),
        GlobalAveragePooling2D(),
        Dropout(0.3),
        Dense(num_classes, activation='softmax')
    ])


def build_resnet_feature_extract(num_classes):
    base = ResNet50(weights='imagenet', include_top=False, input_shape=(*IMG_SIZE, 3))
    base.trainable = False
    inputs = Input(shape=(*IMG_SIZE, 3))
    x = base(inputs, training=False)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    model = Model(inputs, outputs)
    return model


def build_resnet_finetune_last_block(num_classes):
    base = ResNet50(weights='imagenet', include_top=False, input_shape=(*IMG_SIZE, 3))
    # Freeze all, then unfreeze conv5_* block
    base.trainable = False
    for layer in base.layers:
        if layer.name.startswith('conv5_') or layer.name.startswith('bn5_'):
            layer.trainable = True
    inputs = Input(shape=(*IMG_SIZE, 3))
    x = base(inputs, training=True)
    x = GlobalAveragePooling2D()(x)
    x = Dropout(0.4)(x)
    outputs = Dense(num_classes, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

# ---------- TRAIN/REPORT ----------

def compile_and_train(model, train_ds, val_ds, epochs=3, lr=1e-4):
    model.compile(optimizer=Adam(lr), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    history = model.fit(train_ds, validation_data=val_ds, epochs=epochs, verbose=1)
    return history
````

````python
# ========== Variant 1: Scratch CNN ==========
train_scratch, val_scratch, class_names = make_scratch_ds()
num_classes = len(class_names)
model_scratch = build_scratch_cnn(num_classes)
compile_and_train(model_scratch, train_scratch, val_scratch, epochs=5, lr=3e-4)

# quick eval
loss, acc = model_scratch.evaluate(val_scratch, verbose=0)
print(f"Scratch CNN val acc: {acc:.3f}")
````

````python
# ============================================
# CNN CLASSIFICATION - LOAD FROM FOLDER
# ============================================
"""
Folder Structure:
    dataset/
        class_a/
            img1.jpg
            img2.jpg
        class_b/
            img3.jpg
            img4.jpg
            
Labels are AUTOMATIC from folder names (alphabetically):
    class_a → 0
    class_b → 1
"""

# CONFIG
data_dir = 'dataset/'  # CHANGE THIS to your folder path
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 42

# ============ LOAD DATA ============
# 70% training
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.30,    # Hold out 30%
    subset='training',        # This gets 70%
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='int'          # Integer labels (0, 1, 2...)
)

# 30% for val+test
valtest_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.30,
    subset='validation',      # This gets 30%
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    label_mode='int'
)

# Split 30% → 20% val + 10% test
valtest_batches = tf.data.experimental.cardinality(valtest_ds).numpy()
test_batches = valtest_batches // 3
test_ds = valtest_ds.take(test_batches)
val_ds = valtest_ds.skip(test_batches)

# ============ GET CLASS NAMES ============
class_names = train_ds.class_names
num_classes = len(class_names)

print(f"✅ Classes: {class_names}")
print(f"✅ Number of classes: {num_classes}")
print(f"✅ Train batches: {tf.data.experimental.cardinality(train_ds).numpy()}")
print(f"✅ Val batches: {tf.data.experimental.cardinality(val_ds).numpy()}")
print(f"✅ Test batches: {tf.data.experimental.cardinality(test_ds).numpy()}")
````

````python
# ============================================
# CNN - PREPROCESSING (Augmentation + Normalization)
# ============================================

# Normalization layer
rescale = Rescaling(1./255)

# Augmentation (ONLY for training!)
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
])

# Apply to training: Augmentation + Normalization
train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y))
train_ds = train_ds.map(lambda x, y: (rescale(x), y))

# Apply to val/test: ONLY Normalization (NO augmentation!)
val_ds = val_ds.map(lambda x, y: (rescale(x), y))
test_ds = test_ds.map(lambda x, y: (rescale(x), y))

# Optimize pipeline
train_ds = train_ds.prefetch(tf.data.AUTOTUNE)
val_ds = val_ds.prefetch(tf.data.AUTOTUNE)
test_ds = test_ds.prefetch(tf.data.AUTOTUNE)

print("✅ Preprocessing applied!")
````

# 📂 SEGMENTATION DATA LOADING

For segmentation, you need to pair images with their masks manually.

````python
# ============================================
# SEGMENTATION - LOAD IMAGE + MASK PAIRS
# ============================================
"""
Folder Structure:
    root/
        images/
            img1.png
            img2.png
        masks/
            mask1.png  (same order/naming as images)
            mask2.png
"""

# CONFIG
IMG_DIR = 'root/images'   # CHANGE THIS
MASK_DIR = 'root/masks'   # CHANGE THIS
IMG_HEIGHT = 256
IMG_WIDTH = 256
BATCH_SIZE = 8
NUM_CLASSES = 34  # Number of segmentation classes

# ============ GET FILE PATHS ============
img_paths = sorted(glob.glob(os.path.join(IMG_DIR, '*.png')))
mask_paths = sorted(glob.glob(os.path.join(MASK_DIR, '*.png')))

print(f"Found {len(img_paths)} images")
print(f"Found {len(mask_paths)} masks")

# Verify same count
assert len(img_paths) == len(mask_paths), "Mismatch in images and masks!"

# ============ LOAD FUNCTION ============
def load_segmentation_data(img_path, mask_path):
    # Load image
    img = tf.io.read_file(img_path)
    img = tf.image.decode_png(img, channels=3)
    img = tf.image.resize(img, (IMG_HEIGHT, IMG_WIDTH))
    img = tf.cast(img, tf.float32) / 255.0  # Normalize to [0,1]
    
    # Load mask (IMPORTANT: use 'nearest' to preserve class labels!)
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask, channels=1)
    mask = tf.image.resize(mask, (IMG_HEIGHT, IMG_WIDTH), method='nearest')
    mask = tf.cast(mask, tf.int32)
    
    return img, mask

# ============ SPLIT DATA ============
total = len(img_paths)
split_70 = int(total * 0.70)
split_85 = int(total * 0.85)

train_img = img_paths[:split_70]
train_mask = mask_paths[:split_70]

val_img = img_paths[split_70:split_85]
val_mask = mask_paths[split_70:split_85]

test_img = img_paths[split_85:]
test_mask = mask_paths[split_85:]

print(f"Train: {len(train_img)}")
print(f"Val: {len(val_img)}")
print(f"Test: {len(test_img)}")

# ============ CREATE DATASETS ============
def make_seg_dataset(img_list, mask_list):
    ds = tf.data.Dataset.from_tensor_slices((img_list, mask_list))
    ds = ds.map(load_segmentation_data, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(BATCH_SIZE).prefetch(tf.data.AUTOTUNE)
    return ds

train_seg_ds = make_seg_dataset(train_img, train_mask)
val_seg_ds = make_seg_dataset(val_img, val_mask)
test_seg_ds = make_seg_dataset(test_img, test_mask)

print("✅ Segmentation datasets ready!")
````

````python
# ============================================
# SEGMENTATION - VISUALIZE IMAGE + MASK
# ============================================

# Create color palette for visualization
np.random.seed(42)
palette = np.random.randint(0, 255, size=(NUM_CLASSES, 3))

plt.figure(figsize=(15, 5))

for images, masks in train_seg_ds.take(1):
    # Show first sample
    img = images[0].numpy()
    mask = masks[0].numpy().squeeze()
    
    # Original Image
    plt.subplot(1, 3, 1)
    plt.imshow(img)
    plt.title("Image")
    plt.axis('off')
    
    # Mask (raw class IDs)
    plt.subplot(1, 3, 2)
    plt.imshow(mask, cmap='tab20')
    plt.title("Mask (Class IDs)")
    plt.axis('off')
    
    # Mask (colored)
    plt.subplot(1, 3, 3)
    colored_mask = palette[mask]
    plt.imshow(colored_mask.astype(np.uint8))
    plt.title("Mask (Colored)")
    plt.axis('off')

plt.suptitle("Segmentation Sample", fontsize=14)
plt.tight_layout()
plt.show()
````

````python
# ========== Variant 2: ResNet50 Feature Extractor ==========
train_resnet, val_resnet, class_names = make_resnet_ds()
num_classes = len(class_names)
model_resnet_frozen = build_resnet_feature_extract(num_classes)
compile_and_train(model_resnet_frozen, train_resnet, val_resnet, epochs=5, lr=2e-4)

loss, acc = model_resnet_frozen.evaluate(val_resnet, verbose=0)
print(f"ResNet50 frozen val acc: {acc:.3f}")
````

````python
# ========== Variant 3: ResNet50 Fine-tune conv5_x ==========
train_resnet_ft, val_resnet_ft, class_names = make_resnet_ds()
num_classes = len(class_names)
model_resnet_ft = build_resnet_finetune_last_block(num_classes)
compile_and_train(model_resnet_ft, train_resnet_ft, val_resnet_ft, epochs=5, lr=1e-4)

loss, acc = model_resnet_ft.evaluate(val_resnet_ft, verbose=0)
print(f"ResNet50 partial fine-tune val acc: {acc:.3f}")
````
